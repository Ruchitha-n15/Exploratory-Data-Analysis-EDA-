````python
pivot=data.pivot_table(
    index="City",
    columns="Payment",
    values="Total",
    aggfunc="sum"
)

sns.heatmap(pivot,annot=True,fmt='.0f',cmap='YlOrRd')
plt.show()
````

````python
plt.figure(figsize=(8,4))
data.groupby('Product line')['Quantity'].sum().plot(kind='bar',color=plt.cm.Set1.colors)
plt.title('Total Quantity Purchased by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
data["Date"] = pd.to_datetime(data["Date"])
dw = {0:"Mon",1:"Tue",2:"Wed",3:"Thu",4:"Fri",5:"Sat",6:"Sun"}
data["days_of_week"] = data["Date"].dt.dayofweek.map(dw)

data["days_of_week"].value_counts().plot(
    kind="bar",
    color=plt.cm.Set2.colors,
    figsize=(8,5),
    edgecolor='black'
)
plt.title("Number of Sales per Day of the Week", fontsize=14, fontweight='bold')
plt.xlabel("Day of the Week", fontsize=12)
plt.ylabel("Number of Sales", fontsize=12)
plt.xticks(rotation=0)
plt.show()
````
